```python
import spot
spot.setup(show_default='.abr')
```

```python
f = spot.formula('GFa <-> GFb'); f
```

```python
f.translate()
```

```python
def implies(f, g):
    f = spot.formula(f)
    g = spot.formula_Not(spot.formula(g))
    return spot.product(f.translate(), g.translate()).is_empty()
def equiv(f, g):
    return implies(f, g) and implies(g, f)
```

```python
equiv('a U (b U a)', 'b U a')
```

```python
equiv('!(a U b)', '!a U !b')
```
